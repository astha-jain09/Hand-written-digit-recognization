```python
df=pd.read_csv(r"C:\Users\astha\Desktop\Data Science\project\train.csv").as_matrix()
df
```

```python
f=DecisionTreeClassifier()

#testing datasets
xtrain=df[0:21000,1:]
train_label=df[0:21000,0]

f.fit(xtrain,train_label)

#testing data
xtest=df[21000:,1:]
actual_label=df[21000:,0]

dt=xtest[8]
dt.shape=(28,28)
pt.imshow(dt,cmap='gray')
print(f.predict([xtest[8]]))
pt.show()
```

```python
#Accuracy
p=f.predict(xtest)
count=0
for i in range (0,21000):
    count+=1 if p[i]==actual_label[i] else 0
print("Accuracy=",(count/21000)*100)
```
